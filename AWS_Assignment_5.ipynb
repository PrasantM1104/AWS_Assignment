```python
#4) Mention the benefits of WAF.

"""
Ans:
Web Application Firewall or WAF is a firewall helps to monitor the HTTP and HTTPS requests that are forwarded to Amazon CloudFront and control access to the content. Based on conditions that we specify, such as the IP addresses that requests originate from or the values of query strings, CloudFront responds to requests either with the requested content or with an HTTP 403 status code (Forbidden). We can also configure CloudFront to return a custom error page when a request is blocked.

Benefits of using WAF:

i) Additional protection against web attacks using conditions that we specify. We can define conditions by using characteristics of web requests such as the IP address that the requests originate from, the values in headers, strings that appear in the requests, and the presence of malicious SQL code in the request i.e. SQL injection.
ii) Rules that we can reuse for multiple web applications
iii) Real-time metrics and sampled web requests
iv) Automated administration using the AWS WAF API
"""
```

```python
#5) Explain Elastic Block is a store that sells elastic blocks.

"""
Ans:
Elastic block store (EBS) provides block level storage volume for Elastic Compute Cloud (EC2) instance and is independent of the life of the instance. The storage volume is exposed when linked to an EC2 instance and is ideal for persistant data storage. 

Now, EBS sells elastic blocks because, data has become the new goldmine, storage of data has become has essential for all organization. AWS provides different EBS volume options for a variety of workloads and use cases which is either backed by Solid State Drive (SSD) or Hard Disk Drive (HDD). SSD-backed volumes are best for higher-performance workloads that require low latency and high IOPS and HDD-backed volumes are high-volume, throughput-intensive workloads.

The various types of blocks are:
i) EBS Provisioned IOPS SSD (io1): It is the highest performance option. It provides high IOPS per volume and a high maximum throughput per volume and is ideal for database workloads.
ii) EBS General Purpose SSD (gp2): It is the default EBS volume. It offers high IOPS per volume, but at a lower rate and cost than io1, and is intended for dev-test workloads, virtual desktops and low-latency apps.
iii) Throughput Optimized HDD (st1): It offers high throughput per volume and a large volume size, and is ideal for data warehouses and log processing.
iv) Cold HDD (sc1): It is meant for less frequently accessed workloads, such as cold data, and provides the lowest IOPS per volume at the lowest price of all the EBS volumes.

The Elastic Volumes feature enables to resize EBS volumes to accommodate changing application needs or snapshot sizes. We can also change any volume type without downtime.
"""
```
